## Getting Data

```python
path = "data_basis/"
data = parse_ranks(path)

data
```

```python
plot_ranks(path)
```

```python
plot_ranks(path, show_useful = True)
```

# Classification in the M-2-2 Scenario

```python
Ratio = []
M = []

for k in data.keys():
    if k[1:] == '21':
        print(k, data[k])
        M.append(int(k[0]))
        Ratio.append(data[k]['ranks'][1]/data[k]['ranks'][0])
        
Ratio = [r for _,r in sorted(zip(M,Ratio))]
M = sorted(M)
```

```python
fig = plt.figure(figsize=(17, 8))

plt.plot(M, Ratio, linewidth = 3, zorder=1)
plt.scatter(M, Ratio, zorder=2)

plt.xlabel('Number of measurements', fontsize=20)
plt.ylabel('Ratio of ranks', fontsize=20)
plt.tick_params(axis='both', labelsize=18)
plt.show()
```
