## Cycle Graph MSO

```python
index = []
for n in range(2,7):
    key = str(n)+'22'
    index.append(1 - data[key]['ranks'][0]/data[key]['ranks'][1])

fig = plt.figure(figsize=(17, 8))
plt.plot(range(2,7), index, linewidth = 3, zorder=1)
plt.scatter(range(2,7), index, zorder=2)

plt.xticks(range(2,7))
plt.xlabel('N', fontsize=20)
plt.ylabel('Index', fontsize=20)
plt.tick_params(axis='both', labelsize=18)
plt.show()
```
