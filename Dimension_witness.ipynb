# Testing via SDP

```python
eta = cp.Variable((1, 1))
alpha = cp.Variable((len(X_basis_3d), 1), complex=True)
Gamma = cp.Variable(X_basis_3d[0].shape)

constraints = [Gamma >> 0]
constraints += [
    Gamma == sum([alpha[i]*X_basis_3d[i] for i in range(len(X_basis_3d))])
]
for i in range(1,len(X)):
    constraints += [
        eta*X[i,i] + (1-eta)/2 == Gamma[i,i] 
    ]

prob = cp.Problem(cp.Maximize(eta),
                  constraints)

prob.solve(verbose=True)
```
