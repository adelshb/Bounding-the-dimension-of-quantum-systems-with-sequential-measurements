# Classification in the M-2-2 Scenario

```python
Ratio = []
M = []

for mlo in data.keys():
    if mlo[1:3] == '22' and mlo[-1] == '1' and len(data[mlo]['LI num'])>1:
        print(mlo, data[mlo])
        M.append(int(mlo[0]))
        Ratio.append(data[mlo]['LI num'][1]/data[mlo]['LI num'][0])
        
Ratio = [r for _,r in sorted(zip(M,Ratio))]
M = sorted(M)
```

```python
fig = plt.figure(figsize=(17, 8))

plt.plot(M, Ratio, linewidth = 6, zorder=1)
plt.scatter(M, Ratio, s=200, zorder=2)

plt.xlabel('Number of measurements', fontsize=29)
plt.ylabel('Ratio of number of LI moment matrices', fontsize=29)
plt.tick_params(axis='both', labelsize=26)
plt.xticks(M)
plt.show()
```
